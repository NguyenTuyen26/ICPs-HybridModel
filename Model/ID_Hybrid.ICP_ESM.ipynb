```python
import numpy as np
X_train = np.load("/content/drive/MyDrive/ICPS/new_train_X_esm.npy")
y_train = np.load("/content/drive/MyDrive/ICPS/new_train_y_esm.npy")
X_test = np.load("/content/drive/MyDrive/ICPS/new_test_X_esm.npy")
y_test = np.load("/content/drive/MyDrive/ICPS/new_test_y_esm.npy")
```

```python
from sklearn.metrics import confusion_matrix, matthews_corrcoef, accuracy_score, roc_auc_score
from keras.models import load_model

model = load_model("/content/drive/MyDrive/ICPS/ICP_CNN_BiLSTM.h5")

y_pred_prob = model.predict([X_test, X_test])
y_pred = (y_pred_prob > 0.5).astype(int).flatten()

cm = confusion_matrix(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_pred_prob)

sn = cm[1][1] / (cm[1][1] + cm[1][0]) if (cm[1][1] + cm[1][0]) > 0 else 0
sp = cm[0][0] / (cm[0][0] + cm[0][1]) if (cm[0][0] + cm[0][1]) > 0 else 0

# In kết quả
print("Confusion Matrix:\n", cm)
print(f"✅ ACC: {acc:.4f}")
print(f"✅ MCC: {mcc:.4f}")
print(f"✅ SN (Sensitivity): {sn:.4f}")
print(f"✅ SP (Specificity): {sp:.4f}")
print(f"✅ AUC: {auc:.4f}")
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

y_pred_prob = model.predict([X_test, X_test]).flatten()
y_true = np.array(y_test).flatten()

# Tính fpr, tpr
fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
roc_auc = auc(fpr, tpr)

# Vẽ biểu đồ ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)', fontsize=12)
plt.ylabel('True Positive Rate (TPR)', fontsize=12)
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=14)
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
